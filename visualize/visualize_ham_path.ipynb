```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
def parse_graph(filename):
    f = open(filename, "r")
    size = int(f.readline())
    vertices = []
    for line in f:
        x,y = line.split()
        x = float(x)
        y = float(y)
        vertices.append((x, y))
    return vertices
```

```python
def parse_tour(filename):
    f = open(filename, "r")
    tour_str = f.readline()
    return np.array([int(city) for city in tour_str.split(',')[:-1]])
```

```python
def draw_tour(vertices : np.array, tour : np.array):
    fig, ax = plt.subplots(figsize = (6, 6))
    ax.scatter(x = vertices[:,tour][0], y = vertices[:,tour][1])
    ax.plot(vertices[:,tour][0], vertices[:,tour][1])
    ax.set(
        title = f'Гамильтонов цикл, {len(vertices[0])} вершин',
    )
    plt.show()
```

```python
vertices = np.array(parse_graph('../genetic_algorithm/examples/example_s30.txt')).T
tour = parse_tour('../genetic_algorithm/solves/example_s30_solve.txt')
draw_tour(vertices, tour)
#draw_tour(vertices, np.array([0,22,14,29,15,16,19,4,26,6,5,21,13,10,23,24,18,17,8,20,28,3,9,25,2,12,11,7,1,27,0]))
```
